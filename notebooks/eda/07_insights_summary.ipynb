# Insights

This notebook summarizes the main analytical and business insights derived from
descriptive statistics, visualization and distribution fitting of NYC ride demand.

## Demand structure

Ride demand at the zone–hour level is highly right-skewed. 

Most zone–hour combinations have very low demand, while a small fraction of
observations correspond to extremely high-demand periods. The median demand is low, but the upper tail is very heavy, indicating that
a limited number of zones and hours generate a disproportionate share of trips.

## Temporal patterns

Demand follows a strong and regular daily cycle.

The highest average demand occurs in the early evening, with a clear peak around 18:00. Demand increases steadily during the morning, stabilizes during the day, and drops sharply during late night hours. Weekly patterns are also present, with systematic differences between weekdays and weekends.

## Spatial concentration

Ride demand is highly concentrated in a small number of city zones.

A few central zones dominate average hourly demand, while most zones exhibit consistently low activity. This spatial imbalance indicates that demand modeling and operational planning cannot rely on city-wide averages and must account for strong local effects.

## Variability and operational risk

Even within high-demand zones, demand varies substantially from hour to hour.

Extreme peaks occur rarely but represent significant operational stress for ride-hailing platforms, as they can lead to long waiting times and unmet demand if supply is not positioned in advance. This variability makes short-term forecasting particularly important for
operational decision making.


## Distributional properties

Ride demand exhibits clear over-dispersion, with variance significantly larger than the mean.

A Poisson distribution provides a poor approximation due to its inability to capture the heavy right tail. Negative Binomial and lognormal distributions provide a substantially better fit, indicating that demand is both over-dispersed and strongly skewed.

## Implications for modeling

The strong skewness and over-dispersion suggest that standard linear models with Gaussian assumptions are not appropriate for raw demand values.

Future modeling stages should consider:
- transformations of the target variable
- probabilistic models that explicitly handle over-dispersion
- zone-specific effects and clustering
- separate treatment of peak and off-peak periods

## Implications for business and operations

Reliable short-term demand prediction is critical for efficient driver allocation, especially during peak hours and in high-demand zones.

Operational decisions should prioritize:
- proactive driver positioning in a small number of high-impact zones
- dynamic supply planning during evening peaks
- targeted monitoring of zones that exhibit frequent demand spikes

A city-wide strategy alone would fail to capture the highly localized nature of demand.

## Conclusion

This phase confirms that NYC ride demand is highly heterogeneous across both time and space. These findings justify the next project stages, which focus on inferential statistics, zone segmentation and predictive modeling to better explain and forecast this complex demand behavior.
